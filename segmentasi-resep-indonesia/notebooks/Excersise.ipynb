## ðŸ” Analisis Dataset untuk Ketahanan Pangan dan Makanan Favorit

### **1. Zomato Data (Makanan Favorit)**

#### âœ… **Kelebihan**:
- `dish_liked`: Bisa digunakan untuk melihat makanan favorit di Indonesia berdasarkan review pelanggan.
- `votes` & `rate`: Bisa membantu dalam mengukur popularitas makanan.
- `cuisines`: Bisa mengidentifikasi kategori makanan yang paling sering dikonsumsi.
- `reviews_list`: Bisa digunakan untuk analisis sentimen makanan (positif/negatif).

#### âš  **Kekurangan**:
- `dish_liked` hanya memiliki **23.639 dari 51.717** data â†’ Banyak data kosong.
- Tidak ada informasi tentang **asal makanan** atau keterkaitan dengan produksi pangan.
- Data kemungkinan besar berasal dari kota besar, bukan seluruh Indonesia.

#### ðŸ›  **Solusi**:
- Gunakan `cuisines` dan `reviews_list` untuk mendapatkan gambaran umum tentang makanan yang paling disukai.
- Lakukan analisis sentimen untuk mengetahui apakah makanan sehat lebih disukai dibanding makanan tidak sehat.

---

### **2. Production Data (Ketahanan Pangan)**

#### âœ… **Kelebihan**:
- Memuat produksi berbagai bahan pangan seperti **beras, gandum, kentang, ayam, pisang, dll**.
- Bisa digunakan untuk melihat **tren produksi** selama beberapa tahun (`Year`).
- Mampu menghubungkan **produksi vs konsumsi makanan** dari dataset lain.

#### âš  **Kekurangan**:
- Tidak ada informasi konsumsi langsung oleh masyarakat, hanya produksi.
- Tidak ada data keterkaitan dengan pola makan masyarakat dari Zomato.

#### ðŸ›  **Solusi**:
- Hubungkan produksi bahan makanan dengan jumlah restoran yang menyajikan makanan berbahan dasar tersebut (dari **Zomato Data**).
- Analisis korelasi antara **produksi pangan vs makanan yang paling disukai**.

---

### **3. Nutrition Data (Kandungan Gizi)**

#### âœ… **Kelebihan**:
- Memiliki **kalori, protein, lemak, dan karbohidrat** â†’ Bisa digunakan untuk analisis gizi makanan favorit.
- Bisa dikombinasikan dengan Zomato untuk melihat apakah makanan yang paling disukai memiliki **nutrisi seimbang atau tidak**.

#### âš  **Kekurangan**:
- Tidak ada informasi tentang **berapa banyak makanan ini dikonsumsi di Indonesia**.
- Tidak ada informasi keterkaitan dengan produksi pangan.

#### ðŸ›  **Solusi**:
- Hubungkan dengan **Zomato Data** untuk melihat **apakah makanan yang paling disukai tinggi/nutrisi rendah**.
- Buat **skor kesehatan makanan** berdasarkan data kalori, protein, lemak, dan karbohidrat.

---

### **4. Recipe Data (Resep Makanan)**

#### âœ… **Kelebihan**:
- Bisa digunakan untuk mengetahui **bahan makanan yang paling sering digunakan** (`Ingredients`).
- Bisa dihubungkan dengan **Production Data** untuk melihat apakah bahan makanan yang banyak diproduksi juga sering digunakan dalam resep.

#### âš  **Kekurangan**:
- Tidak ada informasi popularitas selain `Loves`, yang mungkin tidak mencerminkan konsumsi nyata.
- Tidak ada informasi lokasi, apakah makanan ini umum di Indonesia atau tidak.

#### ðŸ›  **Solusi**:
- Hubungkan dengan **Production Data** untuk melihat apakah bahan makanan yang banyak diproduksi juga banyak digunakan dalam resep.
- Lihat apakah makanan dengan bahan tertentu memiliki **kandungan gizi yang baik** berdasarkan **Nutrition Data**.

---
